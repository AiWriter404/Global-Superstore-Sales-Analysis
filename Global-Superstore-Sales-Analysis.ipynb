# **Step 1 — Load Dataset**

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/kaggle/input/datasets/muhammadumer7804/store-dataset/Global_Superstore2.csv', encoding='latin1')

print("Shape:", df.shape)
df.head()
```

# **Step 3 — Dataset Preview & Basic Info**

```python
df.head()
```

```python
df.info()
```

# **Step 4 — Missing Values Check (Percentage)**

```python
(df.isnull().sum() / len(df) * 100).sort_values(ascending=False)
```

# **Step 5 — Drop Postal Code Column**

```python
df = df.drop('Postal Code', axis=1)

print(df.shape)
print(df.isnull().sum().sum())
```

# **Step 6 — Convert Date Columns to Datetime**

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d-%m-%Y')

print(df[['Order Date', 'Ship Date']].dtypes)
print(df[['Order Date', 'Ship Date']].head())
```

# **Step 7 — Extract Year and Month from Order Date**

```python
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month

print(df[['Order Date', 'Order Year', 'Order Month']].head())
print(df['Order Year'].unique())
```

# **Step 8 — Overall Business Summary**

```python
print("Total Sales:", df['Sales'].sum())
print("Total Profit:", df['Profit'].sum())
print("Total Orders:", df['Order ID'].nunique())
print("Average Discount:", df['Discount'].mean())
print("Overall Profit Margin (%):", (df['Profit'].sum() / df['Sales'].sum()) * 100)
```

# **Step 9 — Sales & Profit by Category**

```python
category_summary = df.groupby('Category')[['Sales', 'Profit']].sum().sort_values('Profit', ascending=False)
category_summary['Profit Margin (%)'] = (category_summary['Profit'] / category_summary['Sales']) * 100

print(category_summary)
```

# **Step 10 — Visualize Category Performance**

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.barplot(x=category_summary.index, y=category_summary['Sales'], ax=axes[0], palette='Blues_d')
axes[0].set_title('Total Sales by Category')
axes[0].set_ylabel('Sales ($)')

sns.barplot(x=category_summary.index, y=category_summary['Profit'], ax=axes[1], palette='Greens_d')
axes[1].set_title('Total Profit by Category')
axes[1].set_ylabel('Profit ($)')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.barplot(x=category_summary.index, y=category_summary['Sales'], hue=category_summary.index, ax=axes[0], palette='Blues_d', legend=False)
axes[0].set_title('Total Sales by Category')
axes[0].set_ylabel('Sales ($)')

sns.barplot(x=category_summary.index, y=category_summary['Profit'], hue=category_summary.index, ax=axes[1], palette='Greens_d', legend=False)
axes[1].set_title('Total Profit by Category')
axes[1].set_ylabel('Profit ($)')

plt.tight_layout()
plt.show()
```
